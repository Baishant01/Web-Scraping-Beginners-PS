# This code file aims to extract all the stock broker's list which came under Nepse within 2025-Aug-10


```python
# requirements
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Assigniing the url

webpage_request = requests.get("https://merolagani.com/BrokerList.aspx")
webpage_request
```

```python
# Parsing the HTML content using BeautifulSoup
# This will create a BeautifulSoup object that allows us to navigate the HTML structure

soup = BeautifulSoup(webpage_request.text, 'lxml')
webpage_html = soup.prettify()
```

```python
# Extracting the broker list table from the HTML
# This will find the first table in the HTML, which contains the broker information
broker_list_table = soup.find_all('table')[0]
```

```python
# Extracting the column data from the broker list table
# This will find all the rows in the table, which contain the broker information

table_titles_list = soup.find_all('th')
table_titles = [title.text.strip() for title in table_titles_list]
table_titles
```

```python
column_data = broker_list_table.find_all('tr')
```

```python
all_rows = []

for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    all_rows.append(individual_row_data)

# Create DataFrame in one go
df = pd.DataFrame(all_rows, columns=table_titles)  # Add column names if known
pd.set_option('display.max_rows', None)
df
```

```python
# Saving the DataFrame to a CSV file
df.to_csv(r'Brokers List.csv', index = False)
```
